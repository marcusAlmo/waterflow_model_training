## Data Exploration of the Tabular Consumption Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df_consumption = pd.read_csv("../../dataset/clean/production/tabular/consumption_consolidated.csv")
df_consumption.head()
```

```python
df_consumption.isnull().sum()
```

### Disconnected consumers

```python
df_disconnected = df_consumption[df_consumption["Connection Status"] == "Disconnected"]
df_disconnected.head()
```

```python
print("="*20, "Printing df_disconnected.describe()", "="*20)
df_disconnected.describe()
```

```python
print("="*20, "Printing number of nulls in df_disconnected", "="*20)
df_disconnected.isnull().sum()
```

Based on the number of nulls, there are 23 records whose account name is missing. 

```python
# ========== CHECKING THE RECORDS WITH MISSING ACCOUNT NAMES =========
df_disconnected[df_disconnected["Account Name"].isnull()]
```
