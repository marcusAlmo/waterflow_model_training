# Cleaning of consolidated consumption dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df_consumption = pd.read_csv("../../dataset/clean/production/tabular/consumption_consolidated.csv")
df_consumption.head()
```

```python
df_consumption.isnull().sum()
```

### Dropping Control Number duplicates

```python
# Keep the first instance of the duplicate in df_consumption
df_consumption = df_consumption[~df_consumption.duplicated(subset=["Control Number"], keep="first")]
df_consumption.isnull().sum()
```

```python
# Display records with missing district
df_consumption[df_consumption["District"].isnull()]
```

### Filling in missing account names

#### V1 - 8/27/25

```python
df_account_names = pd.read_csv("../../dataset/clean/production/tabular/consolidated_readings_with_account_names.csv")

# Create a mapping from Control Number to Account Name in df_account_names, keeping the first occurrence
control_to_name = df_account_names.drop_duplicates(subset="Control Number").set_index("Control Number")["Account Name"]

# Update Account Name in df where Control Number matches
df_consumption["Account Name"] = df_consumption["Control Number"].map(control_to_name).combine_first(df_consumption["Account Name"])

# Display head
df_consumption.head()
```

```python
df_consumption[df_consumption["Account Name"].isnull()]
```
